## 🔧 BLOCK 1 — Install & Import Libraries

```python
# Install required library
!pip install plotly --quiet

# Import everything we need
import pandas as pd
import numpy as np
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
from IPython.display import display, HTML

print("✅ All libraries loaded successfully!")
```

---
## 📅 BLOCK 2 — Create Sample Financial Data
This generates 12 months of realistic financial data for the dashboard.

```python
# Create 12 months of data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Revenue, Expenses and Profit data (in thousands ₹ or $)
revenue  = [120, 135, 128, 150, 162, 175, 168, 190, 185, 210, 225, 240]
expenses = [ 95, 100,  98, 110, 115, 120, 118, 130, 125, 140, 148, 155]
profit   = [r - e for r, e in zip(revenue, expenses)]

# Category-wise spending breakdown
categories = ['Salaries', 'Marketing', 'Operations', 'R&D', 'Admin']
spending   = [45, 20, 18, 12, 5]

# Stock price simulation
np.random.seed(42)
stock_price = [100]
for _ in range(11):
    change = np.random.uniform(-5, 8)
    stock_price.append(round(stock_price[-1] + change, 2))

# Summary KPIs
total_revenue  = sum(revenue)
total_expenses = sum(expenses)
total_profit   = sum(profit)
profit_margin  = round((total_profit / total_revenue) * 100, 1)

print("✅ Data created!")
print(f"   Total Revenue  : {total_revenue}K")
print(f"   Total Expenses : {total_expenses}K")
print(f"   Total Profit   : {total_profit}K")
print(f"   Profit Margin  : {profit_margin}%")
```

## 📋 BLOCK 3 — Show KPI Summary Cards
Displays the key financial numbers at the top of the dashboard.

```python
# Display styled KPI cards using HTML
kpi_html = f"""
<style>
  .kpi-container {{
    display: flex; gap: 16px; flex-wrap: wrap; font-family: 'Segoe UI', sans-serif;
  }}
  .kpi-card {{
    flex: 1; min-width: 180px; padding: 20px 24px;
    border-radius: 12px; color: white;
    box-shadow: 0 4px 15px rgba(0,0,0,0.15);
  }}
  .kpi-label {{ font-size: 13px; opacity: 0.85; margin-bottom: 6px; }}
  .kpi-value {{ font-size: 28px; font-weight: 700; }}
  .kpi-sub   {{ font-size: 12px; opacity: 0.75; margin-top: 4px; }}
</style>
<div class="kpi-container">
  <div class="kpi-card" style="background: linear-gradient(135deg, #1a73e8, #0d47a1);">
    <div class="kpi-label">💰 Total Revenue</div>
    <div class="kpi-value">{total_revenue}K</div>
    <div class="kpi-sub">Full Year 2024</div>
  </div>
  <div class="kpi-card" style="background: linear-gradient(135deg, #e53935, #b71c1c);">
    <div class="kpi-label">📤 Total Expenses</div>
    <div class="kpi-value">{total_expenses}K</div>
    <div class="kpi-sub">Full Year 2024</div>
  </div>
  <div class="kpi-card" style="background: linear-gradient(135deg, #2e7d32, #1b5e20);">
    <div class="kpi-label">📈 Net Profit</div>
    <div class="kpi-value">{total_profit}K</div>
    <div class="kpi-sub">Full Year 2024</div>
  </div>
  <div class="kpi-card" style="background: linear-gradient(135deg, #6a1b9a, #4a148c);">
    <div class="kpi-label">🎯 Profit Margin</div>
    <div class="kpi-value">{profit_margin}%</div>
    <div class="kpi-sub">Full Year 2024</div>
  </div>
</div>
"""
display(HTML(kpi_html))
print("✅ KPI cards shown above!")
```

## 📈 BLOCK 4 — Revenue vs Expenses Line Chart
Shows monthly trend of income and spending.

```python
fig1 = go.Figure()

# Revenue line
fig1.add_trace(go.Scatter(
    x=months, y=revenue,
    name='Revenue',
    mode='lines+markers',
    line=dict(color='#1a73e8', width=3),
    marker=dict(size=8),
    fill='tozeroy',
    fillcolor='rgba(26,115,232,0.1)'
))

# Expenses line
fig1.add_trace(go.Scatter(
    x=months, y=expenses,
    name='Expenses',
    mode='lines+markers',
    line=dict(color='#e53935', width=3, dash='dash'),
    marker=dict(size=8)
))

fig1.update_layout(
    title='📈 Monthly Revenue vs Expenses (in K)',
    xaxis_title='Month',
    yaxis_title='Amount (K)',
    hovermode='x unified',
    plot_bgcolor='#f8f9fa',
    paper_bgcolor='white',
    height=400,
    legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1)
)

fig1.show()
print("✅ Chart 1 displayed!")
```

---
## 📊 BLOCK 5 — Monthly Profit Bar Chart
Shows how much profit was made each month.

```python
# Color bars: green if profit is high, yellow if low
bar_colors = ['#2e7d32' if p >= 30 else '#f9a825' for p in profit]

fig2 = go.Figure(go.Bar(
    x=months,
    y=profit,
    marker_color=bar_colors,
    text=[f"{p}K" for p in profit],
    textposition='outside'
))

fig2.update_layout(
    title='📊 Monthly Net Profit (Green = High, Yellow = Low)',
    xaxis_title='Month',
    yaxis_title='Profit (K)',
    plot_bgcolor='#f8f9fa',
    paper_bgcolor='white',
    height=400
)

fig2.show()
print("✅ Chart 2 displayed!")
```

---
## 🥧 BLOCK 6 — Expense Breakdown Pie Chart
Shows where the money is being spent.

```python
fig3 = go.Figure(go.Pie(
    labels=categories,
    values=spending,
    hole=0.4,  # Makes it a donut chart
    marker=dict(colors=['#1a73e8','#e53935','#2e7d32','#f9a825','#6a1b9a']),
    textinfo='label+percent',
    hovertemplate='%{label}: %{value}% <extra></extra>'
))

fig3.update_layout(
    title='🥧 Expense Category Breakdown',
    paper_bgcolor='white',
    height=420,
    annotations=[dict(text='Spend', x=0.5, y=0.5, font_size=16, showarrow=False)]
)

fig3.show()
print("✅ Chart 3 displayed!")
```

---
## 📉 BLOCK 7 — Stock Price Trend
Simulates a stock price movement over the year.

```python
fig4 = go.Figure()

fig4.add_trace(go.Scatter(
    x=months,
    y=stock_price,
    mode='lines+markers',
    name='Stock Price',
    line=dict(color='#6a1b9a', width=3),
    marker=dict(size=9, symbol='diamond'),
    fill='tozeroy',
    fillcolor='rgba(106,27,154,0.1)'
))

# Add a horizontal line at starting price
fig4.add_hline(
    y=stock_price[0],
    line_dash='dot',
    line_color='gray',
    annotation_text=f'Start: {stock_price[0]}',
    annotation_position='top left'
)

fig4.update_layout(
    title='📉 Stock Price Trend (Simulated)',
    xaxis_title='Month',
    yaxis_title='Price (₹ / $)',
    plot_bgcolor='#f8f9fa',
    paper_bgcolor='white',
    height=400
)

fig4.show()
print("✅ Chart 4 displayed!")
```

---
## 📋 BLOCK 8 — Monthly Data Table
Shows all data in a clean readable table.

```python
# Create a DataFrame (table)
df = pd.DataFrame({
    'Month'   : months,
    'Revenue (K)' : revenue,
    'Expenses (K)': expenses,
    'Profit (K)'  : profit,
    'Margin %'    : [round((p/r)*100, 1) for p, r in zip(profit, revenue)]
})

# Style the table
styled = df.style \
    .background_gradient(subset=['Revenue (K)'], cmap='Blues') \
    .background_gradient(subset=['Expenses (K)'], cmap='Reds') \
    .background_gradient(subset=['Profit (K)'], cmap='Greens') \
    .set_properties(**{'text-align': 'center', 'font-size': '14px'}) \
    .set_table_styles([{
        'selector': 'th',
        'props': [('background-color', '#1a73e8'), ('color', 'white'),
                  ('font-size', '14px'), ('padding', '10px')]
    }])

display(styled)
print("✅ Data table shown above!")
```
